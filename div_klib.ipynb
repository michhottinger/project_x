```python
df = pd.read_csv('divorce.csv', sep=';')
klib.missingval_plot(df)
```

```python
df.head(5)
```

```python
#data Cleaning
df_cleaned = klib.data_cleaning(df)
```

```python
#correlation plot
klib.corr_plot(df_cleaned, annot=False)
```

```python
klib.corr_plot(df_cleaned, split='pos', annot=False)
```

```python
#There appears to not be any neg corr present in the cleaned data.

#klib.corr_plot(df_cleaned, split='neg', annot=False)
```

```python
#correlation with target feature on interest
klib.corr_plot(df_cleaned, target='atr23')
```

```python
klib.corr_mat(df_cleaned)
```

```python
#dist plot for columns of interest
klib.dist_plot(df_cleaned['atr23'])
```

```python
#categorical plot
klib.cat_plot(df_cleaned, figsize=(50, 15))
```
